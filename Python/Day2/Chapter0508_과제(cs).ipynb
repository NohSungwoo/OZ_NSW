[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


파이썬은 대표적인 동적 언어입니다.
동적 언어는 변수의 자료 유형을 프로그램 실행 중에 정하거나 변경하는 언어를 말하는데
이는 할당 시에 자료형이 정해진다는 것을 의미합니다.
자료형이 명시되지 않았기 때문에 변수의 값을 얼마든지 다른 형태로 전환할 수 있습니다.

반면에 자바와 같은 정적 언어는 선언시에 자료형이 정해집니다.
선언시에 명시적으로 자료형이 정해져 있기 때문에 
변수의 값을 개발자가 임의로 바꾸게 되면 에러가 발생하게 됩니다.

동적 할당은 편리하지만 에러를 발생시키기 더 쉽기 때문에
개발자에게 더 많은 주의를 요구하게 됩니다.

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

실수를 표현하는 방식은 크게 고정 소수점 방식과 부동 소수점 방식을 사용하는데
파이썬은 이 중 부동 소수점 방식을 사용합니다.

부동 소수점 방식은 고정 소수점 방식보다 훨씬 많은 범위까지 표현할 수 있지만, 
숫자가 표현할 수 있는 한계를 넘어설 수는 없습니다.
십진법에서 0.33333333...과 같은 소수를 정확하게 표현할 수 없는 것처럼
이진법에서는 0.1이라는 소수가 정확하게 표현될 수 없습니다.

메모리의 한계를 부딪치게 되기 때문에 한정된 비트 수에 맞춰진 근사값을 얻어오게 됩니다.
그래서 실수형 계산 시에는 오차가 발생하게 됩니다.
