[문제1]

list와 tuple의 주된 차이점 은 무엇인가요?

list와 tuple은 서로 공통점이 많은 데이터 타입입니다.
둘 다 여러개의 데이터를 하나로 만들어줄 수 있지만, 
list는 선언된 값들을 수정할 수 있고
tuple은 선언된 값들을 수정할 수 없습니다.

이 차이는 메모리 할당 방식에서 특히 도드라집니다.  
list는 객체가 생성된 후에도 변경이 가능하기 때문에 
메모리 블록에 추가된 여유 메모리를 할당합니다.
하지만 tuple은 값이 변경될 수 없기 때문에 최소한의 메모리만을 할당하게 되죠.
결국 list는 크키가 커질수록 무거워지게 됩니다.

결론적으로 시스템의 부하를 줄이기 위해 tuple이라는 자료형이 생기게 되었다고 할 수 있습니다.

[문제3]

Python의 list는 연결리스트(linked list) 인가요?

파이썬의 list는 dynamic array 로 구현되어 있습니다.
각 요소는 연속된 메모리 위치에 저장되며, 인덱스를 사용하여 접근할 수 있습니다.
내부적으로는 배열을 사용하지만 리스트가 증가할 때마다 배열의 크기를 조절할 수 있습니다.
이는 연결 리스트 (linked list)와는 다릅니다. 연결 리스트는 각 요소가 다음 요소를 가르키는 포인터를 포함하는 노드로 구성됩니다.
만약 연결 리스트를 사용했다면 각 요소에 접근하려면 처음부터 순회해야 합니다.
