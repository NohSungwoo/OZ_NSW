[문제1]

객체지향에 대해서 설명해주세요

객체 지향은 프로그래밍 방식의 하나입니다.
현실 세계를 모방하여 소프트웨어를 개발하는 방법론입니다.

객체 지향의 주요 개념은 이렇습니다.

1. 객체 : 속성과 메소드를 가진 독립적인 단위입니다. 
2. 클래스 : 객체를 생성하는 틀 또는 설계도입니다. 
객체의 속성과 메서드를 정의합니다.
3. 상속 : 기존 클래스의 속성과 메소드를 다른 클래스가 물려받는 것을 의미합니다.
4. 캡슐화 : 객체의 속성과 메소드를 하나로 묶고, 구현 내용의 일부를 외부에서 볼 수 없게 은닉합니다.
5. 다형성 : 하나의 메소드나 클래스가 있을 때 다양한 방법으로 동작하는 것을 말합니다.

객체 지향에는 다섯가지 원칙이 있는데
* 단일 책임의 원칙
* 개방-폐쇄의 원칙
* 리스코프 치환 원칙
* 의존 역전 원칙
* 인터페이스 분리 원칙
이 있습니다.

객체 지향의 장점은
* 코드 재사용률이 높습니다.
* 유지 보수가 쉽습니다.
* 대형 프로젝트에 적합합니다.

단점은
* 상대적으로 처리 속도가 느립니다.
* 용량이 커질 수 있습니다.
* 보다 많은 시간과 노력이 듭니다.

[문제2]

init 메소드에 대해 설명해주세요

init은 클래스 생성자 메소드입니다. 
클래스의 인스턴스를 생성할 때 자동으로 호출되며, 인스턴스가 생성될 때 초기화 작업을 수행합니다.

init은 self를 첫 번째 매개변수로 사용하는데, self를 통해 인스턴스 변수를 초기화합니다.
self 말고 다른 단어를 사용할 수 있지만, 개발 시 혼동을 줄 수 있다.

[문제3]

클래스 상속에 대해 설명해주세요

하나의 클래스가 다른 클래스의 속성과 메소드를 물려받는 것을 뜻합니다.
이를 통해 코드의 중복을 줄이고 재사용성을 높일 수 있습니다.

부모 클래스를 상속받아 자식 클래스가 부모의 속성과 메소드를 사용할 수 있게 됩니다.
필요에 따라 오버라이딩할 수 있습니다.

[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

MRO는 Method Resolution Order의 약자로 메소드의 결정 순서를 의미합니다.
파이썬은 다중 상속을 지원하기 때문에 동일한 이름을 거진 부모 클래스를 다중 상속 받게 되면
죽음의 다이아몬드라는 문제를 발생시킵니다.
같은 메소드 명을 가진 부모가 존재하면 누구의 메소드를 먼저 상속받는지 알 수 없어집니다.

이런 문제를 해결하기 위한 알고리즘을 적용하여 순서를 확인하는 것을 MRO라고 부릅니다.
MRO의 조건은 다음과 같습니다.
1. 자식 클래스를 부모 클래스보다 먼저 확인합니다.
2. 만약 부모 클래스가 둘 이상이라면 나열된 순서대로 확인합니다.
3. 유효한 후보가 둘 이상 있을 경우 첫 번째 부모 클래스부터 선택한다.
4. 부모 클래스는 중복될 수 있지만 서로 교차되면 TypeError를 일으킨다.

[문제5]

Self는 어떤 키워드 인가요?

self는 파이썬에서 클래스의 인스턴스 메소드에서 현재의 인스턴스를 참조하는 데 사용되는 키워드입니다.
self는 메소드가 호출되는 객체에서 자신을 가리킵니다.
파이썬에서는 이를 명시적으로 첫 번째 매개변수로 표기합니다.

[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

TypeError : 잘못된 타입의 연산이나 함수 호출이 일어났을 때 발생합니다.
NameError : 변수 이름을 찾을 수 없을 때 발생합니다.
FileNotFoundError : 존재하지 않는 파일을 열려고 시도하면 발생하는 에러입니다.
