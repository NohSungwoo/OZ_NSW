[문제6]

5번 문제에서 생성한 인스턴스의 이름, 나이, 성별을 출력해주세요.
인스턴스 변수에 접근하여 값을 출력하면 됩니다  : )

출력 결과  
이름 : 김코  
나이 : 30  
코딩레벨 : A  

```python
class Oz : 
    def __init__(self, name, age, grade) :
        self.name = name
        self.age = age
        self.grade = grade
    def student(self, name, age, grade) :
        print(f'이름 : {name}\n나이 : {age},\n코딩레벨 : {grade}입니다.')
coding = Oz("김코", 30, "A")
coding.student(coding.name, coding.age, coding.grade)
```

[뮨제7]

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 : 이름: 김코, 나이: 30, 코딩레벨: A

```python
class Oz : 
    def __init__(self, name, age, grade) :
        self.name = name
        self.age = age
        self.grade = grade
    def who(self, name, age, grade) :
        print(f'이름 : {name}\n나이 : {age},\n코딩레벨 : {grade}입니다.')
coding = Oz("김코", 30, "A")
coding.who(coding.name, coding.age, coding.grade)
```
