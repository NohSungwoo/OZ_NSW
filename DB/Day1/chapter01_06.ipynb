[문제1]

DBMS 특징이 아닌것을 골라주세요

- 데이터 무결성
- 데이터 보안성
- 데이터 일관성
- 데이터 의존성

정답 : 데이터 의존성

데이터 무결성 : 데이터의 정확성, 일관성, 유효성이 유지되는 것을 말한다.

데이터 보안성 : 데이터 보안성은 데이터의 무단 액세스, 변경, 손상으로부터 보호하는 것을 의미한다.

데이터 일관성 : 데이터는 모든 사본 간에 데이터가 일관되어야 하며, DB 내부의 데이터 모델도 일관성을 유지해야 한다.

데이터 의존성 : 데이터 의존성은 데이터 간의 관계를 나타내는 개념이다. 이는 DB 설계 및 모델링 단계에서 고려되는 중요한 요소이지만, 핵심적인 특징은 아니다.

[문제2]

아래 설명에 해당하는 데이터 모델을 골라주세요

```
표 형태로 데이터를 저장하고 SQL을 사용하여 데이터를 조작합니다. 
가장 널리 사용되는 데이터 모델 중 하나이며 RDBMS에서 지원합니다.
```

- 관계형 데이터 모델
- 객체 지향형 데이터 모델
- 계층형 데이터 모델
- 네트워크형 데이터 모델

정답 : 관계형 데이터 모델

관계형 데이터 모델 : 표 형태로 데이터를 저장하고 RDBMS에서 가장 보편적으로 사용되는 데이터 모델이다. 
이 모델에서는 데이터를 테이블 형태로 표현하고, 각 테이블은 행과 열로 구성된다. 

객체 지향형 데이터 모델 : 현실 세계의 객체와 그들 간의 관계를 표현하는 데이터 모델이다.
이 모델은 객체 지향 프로그래밍에서 사용되는 개념을 DB에 적용하여 객체 간의 관계를 나타낸다.

계층형 데이터 모델 : 계층형 데이터 모델은 데이터를 트리 구조로 표현하는 데이터 모델이다.
데이터가 부모 자식 관계를 가지게 되며, 각 레코드는 하나의 부모를 가지고 있다.

네트워크형 데이터 모델 : 네트워크형 데이터 모델은 계층형 데이터 모델의 한계를 극복하기 위해 등장했다.
데이터가 네트워크 구조로 표현되며 하나의 레코드가 여러 개의 부모 레코드를 가질 수 있다.
레코드 간의 관계는 link로 연결되어 있으며, 이를 통해 복잡한 관계를 표현할 수 있지만, 현재는 잘 사용하지 않는 경우가 많다.

SQL : DB에서 데이터를 조작하기 위해 표준화된 언어이다.

[문제6]

대표적인 RDBMS가 아닌것을 골라주세요

- SQLite
- Oracle Database
- PostgreSQL
- Firebase

정답 : Firebase

SQLite : 서버가 아닌 클라이언트 측에 내장된 관계형 DBMS이다.
가볍게 동작하고 별도의 서버가 필요하지 않다. 단일 파일로 구성되어 있어 쉽게 이식할 수 있다.
주로 모바일 애플리케이션과 임베디드 시스템에서 사용된다.

Oracle Database : 오라클이 개발한 대규모 DBMS이다. 가장 인기있는 DBMS 중 하나이며
높은 성능, 확장성, 신뢰성을 제공한다. 주로 대규모 기업에서 많이 사용된다.

PostgreSQL : 오픈 소스 관계형 DBMS로 강력한 기능과 확장성을 제공한다.
ACID트랜잭션, 외래 키, 뷰, 저장 프로시저 등의 고급 기능을 제공하고 다양한 운영체제에서 사용 가능하다.

Firebase : Google이 제공하는 모바일 및 웹 애플리케이션 개발 플랫폼이다.
실시간으로 DB, 인증, 호스팅, 클라우드 저장소 등 다양한 기능을 제공하여 개발자가 쉽고 빠르게 어플리케이션 구축이 가능하다.

- 데이터베이스 사용자 뷰

[문제8]

SQL 언어에 해당하지 않는 것을 골라주세요

- 데이터 처리 언어(DML)
- 데이터 제어 언어(DCL)
- 데이터 연결 언어(TCL)
- 데이터 정의 언어(DDL)

정답 : 데이터 연결 언어(TCL)

데이터 처리 언어(DML) :  SQL에 데이터를 검색, 삽입, 수정하는데 사용되는 언어다. SELECT, INSERT, UPDATE, DELETE가 있다.

데이터 제어 언어(DCL) : SQL에서 DB의 사용권한을 제어하는 언어이다. 주요 명령으로는 GRANT, REVOKE 등이 있다.

데이터 연결 언어(TCL) : DB의 트랜젝션을 제어하는 데 사용되는 언어이다. TCL은 COMMIT, ROLLBACK, SAVEPOINT, RELEASE SAVEPOINT, SET TRANSACTION 등이 있다.

데이터 정의 언어(DDL) : DB의 구조를 정의하는 데 사용되는 언어이다. CREATE, ALTER, DROP등이 있다.
